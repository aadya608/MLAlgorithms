<h2>Confusion Matrix</h2>

<p>
A confusion matrix is a table used to evaluate the performance of a machine learning model on a <b>classification problem.</b>
 It summarizes the number of correct and incorrect predictions made by the model on a set of test data, compared to the actual outcomes.

The confusion matrix is a square matrix with the number of rows and columns equal to the number of classes in the problem. The four cells of the matrix represent the following:

<b>True positives (TP)</b>: the number of instances that are correctly predicted as positive by the model


<b>False positives (FP)</b>: the number of instances that are incorrectly predicted as positive by the model


<b>True negatives (TN)</b>: the number of instances that are correctly predicted as negative by the model


<b>False negatives (FN)</b>: the number of instances that are incorrectly predicted as negative by the model


A confusion matrix can be used to calculate various evaluation metrics such as accuracy, precision, recall, and F1-score. <b>It provides a more detailed analysis of the model's performance than just using a single metric, allowing you to identify which classes the model is performing well on and which ones it is struggling with.</b>

</p>

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
